```python
# Initialize variables and give the spring constant an arbitrary value
v = 0
m = 1
k = 1
t = 0
dt = 0.01
x = -4

# Initialize arrays
time = []
kinetic = []
potential = []
total = []

# Initialize objects
left = box(pos = vector(-5, 0, 0), size = vector(1, 1, 1), mass = m, radius=1)
right = sphere(pos = vector(1, 0, 0), mass = m,  radius=1)
spring = helix(pos = vector(-5, 0, 0), axis = vector(4, 0, 0), length = 8)

while t < 10:
    rate(40)
    right.pos = vector(x, 0, 0)
    spring.axis = vector(x+5, 0, 0)

    # F = ma = kx
    a = (-k*x)/m
    
    # v = v + a*t
    v += a * dt

    # p = p + v*t
    x += v * dt

    # Increment time
    t += dt
    
    # Kinetic and Potential energy calculations
    KE = (m/2)*pow(v, 2)
    PE = (k/2)*pow(x, 2)
    TE = KE + PE
    
    # Append values to list
    time.append(t)
    kinetic.append(KE)
    potential.append(PE)
    total.append(TE)

plt.plot(time, kinetic)
plt.plot(time, potential)
plt.plot(time, total)
plt.title('Energy of System')
plt.xlabel('Time [s]')
plt.ylabel('Energy [J]')
plt.legend(['Kinetic', 'Potential', 'Total'])
```

<h3><font color="black">Crystal Lattice Using Forward Euler</font></h3>

```python
scene = canvas()
```

```python
# Initializing N (the dimensions of the NxNxN "crystal lattice")
N = 3
# Initializing mass and spring constant
mass = 1
k = 1
# Diameter is initialized to 2 and the spacing is double that, this will allow room for "springs"
diameter = 2
spacing = 4
# A dictionary is used to hold our sphere and box objects
spheres = {}
boxes = {}
# Initialize time and time step
t = 0
dt = 0.1
# Initialize energy variables
PE = 0
KE = 0
TE = 0
# Initialize arrays
time = []
kinetic = []
potential = []
total = []
# Center on the middle sphere in the lattice with a scaled adjustment
scene.center = 0.5*(N-1)*vector(1,1,1)
# Initialize variables for calculating average time for respective method
avg = 0
count = 0

# Creating an NxNxN matrix of spheres and boxes, the range was modified in order to make calling of the sphere objects more intuitive
for x in range (1, N+1):
    for y in range (1, N+1):
        for z in range (1, N+1):
            velocity = vector(x, y, z)
            spheres[(x, y, z)] = sphere(pos = vector(x*spacing, y*spacing, z*spacing), velocity = velocity.random(), mass = mass, radius=diameter/2, visible = True)
            boxes[(x, y, z)] = box(pos = vector(x*spacing, y*spacing, z*spacing), velocity = 0, size = vector(diameter, diameter, diameter), visible = False)

# Here we check that all velocities are initialized at random from to hold a value between -1 to 1
# for x in range (1, N+1):
#     for y in range (1, N+1):
#         for z in range (1, N+1):
#             check = spheres[(x, y, z)].velocity
#             print(check)
            
while t < 1200:
    rate(500)
    
    # Here we utilize the explicit/forward euler integration method
    for x in range (1, N+1):
        for y in range (1, N+1):
            for z in range (1, N+1):
                
                # Index objects
                s = spheres[(x, y, z)]
                    
                # Update acceleration, velocity, and position. We subtract a vector for better visualization of the change in energy
                a = (-k*s.pos)/mass-vector(2,0.75,0.5)
                s.velocity += a * dt-vector(2,0.75,0.5)
                starttime = timeit.default_timer()
                #print("The start time is :",starttime)
                s.pos +=  s.velocity*dt-vector(2,0.75,0.5)
                #print("The time difference for the Forward EUler is :", timeit.default_timer() - starttime)
                tempavg = timeit.default_timer() - starttime
                avg += tempavg
                
                # Kinetic and Potential energy calculations
                tempKE = (mass/2)*(pow(s.velocity.x,2)+pow(s.velocity.y,2)+pow(s.velocity.z,2))
                tempPE = (k/2)*(pow(s.pos.x,2)+pow(s.pos.y,2)+pow(s.pos.z,2))
                tempTE = tempKE + tempPE
                KE += tempKE
                PE += tempPE
                TE = KE + PE
                t += dt
                count += 1
            
    average = avg / count
    print("Time average time for Forward Euler is: ", average)
    # Append values to list, since we are outside the for loops we get the energy of the system
    time.append(t)
    kinetic.append(KE)
    potential.append(PE)
    total.append(TE)
    
    # Plot the energy of the system
    plt.plot(time, kinetic)
    plt.plot(time, potential)
    plt.plot(time, total)
    plt.title('Energy of System')
    plt.xlabel('Time [s]')
    plt.ylabel('Energy [J]')
    plt.legend(['Kinetic', 'Potential', 'Total'])
```
